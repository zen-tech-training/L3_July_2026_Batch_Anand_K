```python
from langchain_ollama import ChatOllama

# Built-in tool provided by langchain
from langchain_community.tools import DuckDuckGoSearchResults
import langchain
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
from tavily import TavilyClient
langchain.__version__
from langchain.agents import create_agent
from langchain.tools import tool
import os
from dotenv import load_dotenv, find_dotenv

env_path = find_dotenv()
load_dotenv(env_path)
tavilyApiKey = os.getenv("TAVILY_API_KEY")
model = ChatOllama(model="gpt-oss:120b-cloud")
```

```python
search_tool = DuckDuckGoSearchResults()
response = search_tool.invoke("News about CJP students protest in New Delhi")
print(response)
```

```python
wiki_wrapper = WikipediaAPIWrapper(
    top_k_results=2,
    doc_content_chars_max=1000
)

wikipedia_tool = WikipediaQueryRun(api_wrapper=wiki_wrapper)
response = wikipedia_tool.invoke("Neural Networks")
print(response)
```

```python
tavily_tool = TavilyClient(api_key=tavilyApiKey)
response = tavily_tool.search("What is Artificial Intelligence?")
print(response)
```
